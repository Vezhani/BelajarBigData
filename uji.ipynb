```python
from matplotlib import pyplot as plt
from matplotlib import style

x = [0, 1, 2, 3, 4, 5]
y = [24.27, 23.18, 22.19, 8.41, 7.19, 5.62]

fig, ax = plt.subplots()

ax.bar(x, y, align='center')

ax.set_title('Bahasa Pemograman Favorit (UMSU)')
ax.set_ylabel('Pengguna')
ax.set_xlabel('Bahasa')

ax.set_xticks(x)
ax.set_xticklabels(("Python", "JavaScript", "Java", "Ca", "PHP", "C++"))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/sample_data/gk (1).csv')

provinsi = data['provinsi']
JAMBI = data['gk.makanan.perkotaan.2015.maret']

indeks_terpotong = 7

besaran_terpotong = 0.5

explode = [besaran_terpotong if i == indeks_terpotong else 0 for i in range (len(provinsi))]

fig, ax = plt.subplots(figsize=(10, 8))

ax.pie(JAMBI, labels=provinsi, autopct='%1.1f%%', startangle=90, explode=explode)

ax.set_title('Kesejahteraan Pekerja Indonesia')

plt.axis('equal')

plt.show()
```
